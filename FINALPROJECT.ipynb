## $\text{Node Class}$ <a id='NC'> </a>
[$\text{(Return to Contents)}$](#return)



&ensp;&ensp;&ensp;Nodes, as indicated below, have attributes data, index, next, prev, up, down, right, and left. The directionally named attributes will serve as the 'arrows', or connections, between nodes. 

&ensp;&ensp;&ensp;Also, please note that I will be using operator overloading due to its convenient and comprehensible nature.So that we can use Node objects for all of the structures, it will be initialized with all possibly used attributes, using ```__init__``` and so that we can print each node with the index and the data, we will utilize the ```__str__``` overloader.

````
class Node:
   '''Object with attributes:
   data, index, next, prev, up, down, right, and left.'''
   # if you use index, make sure that it is accounted for:
   #  - incremented appropriately when adding
   #  - decremented appropriately when subtracting
   def __init__(Node,data=None,index=0,Next=None,prev=None,up=None,down=None,right=None,left=None):
       Node.data  = data ;#int,float, string, or some other object (theoretically)
       Node.next  = Next ;
       Node.prev  = prev ;
       Node.up    = up   ;
       Node.down  = down ;
       # mainly for trees and net structure
       Node.left = left;
       Node.right = right;
       Node.index = index;
   def __str__(Node):# Node is used in replacement of self here
       if type(Node.data)==(int or float):# special instructions when Node is a number
           return '(%.d,%.d)' % (Node.index,Node.data)
       else:
           return str((Node.index,Node.data))
````

````
class SLL:# single linked list
   '''A list composed of nodes, each with a single or one way facing arrow.
   Ex.: (Front) -> (A) -> (B) -> (Next)'''
   def __init__(SLL,Front=None,icount=0):#last and Next are used interchangeably here
       Front = Node()
       SLL.front = Front;
       SLL.last = Front; # to start, but it will move as nodes are added
       SLL.icount = icount;
       #Front=None
   def __str__(SLL):# stores all of the nodes as a string to be printed
       temp = SLL.front
       k = ''
       while type(temp)==Node:# !=None didn't work
           k= k+str(temp)
           if type(temp.next)==Node:# !=None didn't work
               k = k+'->'
           temp = temp.next
       return k
   def __add__(SLL,data):
       i = SLL.icount
       temp = SLL.last
       if type(data)!=Node:
         temp.next = Node(data,i+1)
       if type(data)==Node:
         temp.next = data;
         temp.next.index = i+1# always account for the index
       SLL.last = temp.next;
       SLL.icount+=1
       return SLL # so that we can add more than one node in each line when we use the + operator     
   def add(SLL,data):# should I make this data able to be a node
       i = SLL.icount# account for index
       temp = SLL.last
       if type(data)!=Node:
        temp.next = Node(data,i+1)
       if type(data)==Node:
        temp.next = data;
        temp.next.index = i+1
       SLL.last = temp.next;
       SLL.icount+=1
       return SLL# so that we can more than one node in each line when add is called
   # add() can not be called like: sll.add('#')+sll.add('other#'), 
   #                   instead do: sll.add('#'),sll.add('other#')
   def at(SLL,index=None,data=None):# clarify what you are searching by
     temp = SLL.front
     # we need to know what the node is being searched by
     if index!=None and data!=None:
       return "Do not search by index and data at the same time."
     elif index!=None or data!=None:
       while temp!=None:
         if temp.index==index or str(temp.data)==str(data):
            break;
         else:
           temp = temp.next
       if temp==None:
         if index!=None:
           return "No node in the list has index = "+str(index)+"."
         if data!=None:
           return "No node in the list has data = "+str(data)+"."
       else:
         return temp;
     else:
       return "No index or data given, there is nothing to search for.";
   # so that we can print the list as a more organized table(looks nicer)
   def table(SLL):
     data = []
     temp = SLL.front 
     while temp!=None:
       data.append([temp.data])
       temp = temp.next
     return pd.DataFrame(data,columns=["Data"])
````

````
class DLL:# doubly linked list
   '''A list composed of nodes, each with two one-way facing arrows opposite in direction to each other.
   Ex.: (Front) <=> (A) <=> (B) <=> (Next)'''
   def __init__(DLL,Front=None,icount=0):#last and Next are used interchangeably here
       Front = Node()
       #Last = Front;
       DLL.front = Front;
       DLL.front.prev = None # there is nothing left of last
       #DLL.last = Last;
       DLL.last = Front;
       DLL.last.next  = None # there is nothing right of last
       DLL.front.next = DLL.last# Front->Last to start
       DLL.last.prev  = DLL.front# Front<-Last to start
       # also the last and front node are the same Node to start
       #DLL.front = DLL.last # is there a case where this is needed??
       DLL.icount = icount;       
   def __str__(DLL):# stores all of the nodes as a string to be printed
       temp = DLL.front
       k = ''
       while type(temp)==Node:
           k= k+str(temp)
           if type(temp.next)==Node:
               k = k+'<=>'
           temp = temp.next
       return k
   def __add__(DLL,data):
     i = DLL.icount
     temp = DLL.last
     temp.next = Node(data,i+1)
     temp.next.prev = temp;
     DLL.last = temp.next;
     DLL.icount+=1
     return DLL # so that we can add more than one node in each line 
   def add(DLL,data):#addNode to the 'end'(the far right) of the list
      i = DLL.icount
      temp = DLL.last
      temp.next = Node(data,i+1)
      temp.next.prev=temp
      DLL.last = temp.next
      DLL.icount+=1
      return DLL
   def at(DLL,index=None,data=None):# clarify what you are searching by 
     temp = DLL.front
     if index!=None and data!=None:
       return "Do not search by index and data at the same time."
     elif index!=None or data!=None:
       while temp!=None:
         if temp.index==index or str(temp.data)==str(data):
            break;
         else:
           temp = temp.next
       if temp==None:
         if index!=None:
           return "No node in the list has index = "+str(index)+"."
         if data!=None:
           return "No node in the list has data = "+str(data)+"."
       else:
         return temp;
     else:
       return "No index or data given, there is nothing to search for.";
   def table(DLL): 
     data = []
     temp = DLL.front  
     while temp!=None:
       data.append([temp.data])
       temp = temp.next
     return pd.DataFrame(data,columns=["Data"])
# if I add a subtract or delete function I will have to take care of changing all of the indexes respectively
````

````
class CLL:# circularly linked list
    '''A list composed of nodes, each with two one-way facing arrows, oposite in direction to each other, 
    where the last node is doubly connected to the header.
    Ex.: (Header) <=> (A) <=> (B) <=> (C or Header prev)'''
    def __init__(CLL,header=None,icount=0):#last and Next are used interchangeably here
        header = Node()
        CLL.header = header;
        CLL.header.next = CLL.header
        CLL.header.prev = CLL.header
        #header = CLL.header # will this even work??
        CLL.icount = icount
    def __str__(CLL):# stores all of the nodes as a string to be printed
        temp = CLL.header.next#could make this header.prev if prev set up
        k = '|-->'+str(CLL.header)+'\n|      ^         \n|      |         \n|      v         \n';
        while type(temp)==Node and temp!=CLL.header:
            k+='|'
            if temp.next==CLL.header:
              k+='-->'
            if temp.next!=CLL.header:
              k+='   '
            k+=str(temp)+'\n'
            if type(temp.next)==Node:
              if temp.next!=CLL.header:
                k +='|      ^         \n|      |         \n|      v         \n'
            temp = temp.next
        # the if statement below is mainy for the net structure
        if CLL.header.next==CLL.header or CLL.header.data==None:
          k = str(CLL.header)
        return k
    # you can add a node or a string. either way works
    def __add__(CLL,data):
      i = CLL.icount
      if (type(data)!=Node):
        nn= Node(data,i+1)
      if (type(data)==Node):
        nn = data;
        nn.index = i+1
      if CLL.header.data!=None:
        k = CLL.header.prev
        k.next = nn; k.next.prev = k
        nn.next = CLL.header; CLL.header.prev = nn;
        CLL.icount+=1
      else:
        CLL.header.data = nn.data
        CLL.icount-=1;
        del nn;#nn does not need to exist if header is the only node in the CLL
      return CLL # so that we can add mutiple nodes at a time
    def __sub__(CLL,data):
      if type(data)==Node:# so that we can add it as a node or a string or number
        data = data.data;
      # deletes the most recent instance of node with node.data=data
      # starts with the header.prev
      temp = CLL.header.prev;
      while temp!=CLL.header:
        if temp.data==data:
          temp.next.prev = temp.prev;
          temp.next.prev.next = temp.next;
          del temp
          temp = None;# if you comment this out you will see why its necessary
          CLL.icount-=1
          break;
        temp = temp.prev;
      if temp==CLL.header:
        if temp.data==data:
          temp = temp.next
          CLL.header.data = CLL.header.next.data;
          CLL.header.next = CLL.header.next.next;
          CLL.header.next.prev = CLL.header;
          CLL.icount-=1
          del temp;
          temp=None
          #return CLL;
        #elif temp!=None:
          #print(str(data) + " has not been found in any node in the list.\n")
          #return;
          #break;
          #temp=None
      # make sure all of the indexes are correct;
      iFix = CLL.header.next;
      i=1;
      while iFix!=CLL.header: 
        iFix.index=i;
        i+=1;
        iFix = iFix.next;
      return CLL;
    # make a function that deletes all instances of a data
    def add(CLL,data):#addNode to the 'end'(the far right) of the list
        i = CLL.icount
        #temp = CLL.header
        #while type(temp.next)==Node:
          #temp = temp.next
        nn= Node(data,i+1)
        k = CLL.header.prev
        k.next = nn; k.next.prev = k
        nn.next = CLL.header; CLL.header.prev = nn;
        #nn.next.prev=nn
        #nn.prev.next=nn
        #CLL.last = temp.next
        CLL.icount+=1
        return CLL # IF THERE ARE ANY ISSUES CHECK HERE
    def deleteAll(CLL,data):# to delete all occurances of a data member
      if type(data)==Node:
        data = data.data;
      elif type(data)==list or type(data)==tuple:
        for x in range(len(data)):
          if CLL.header.data==data[x]:
            CLL-data[x];
          temp = CLL.header.next;
          while temp!=CLL.header:
            CLL-data[x];
            #continue#do not uncomment I belive it causes infinite loop
            temp = temp.next;
      else:   
        if CLL.header.data==data:
          CLL-data;
        temp = CLL.header.next;
        while temp!=CLL.header:
          CLL-data;
          #continue#do not uncomment I belive it causes infinite loop
          temp = temp.next;
    def at(CLL,index):
      iN = CLL.header
      if index<=CLL.icount:# if the index being searched for is within bounds
        while iN.index!=index:
          iN = iN.next
      if index!=0 and iN==CLL.header:
        return ' Node of index '+str(index)+' does not exist in this list.'
      else:
        return iN;
````

````
class Net:# single linked list
    '''Structure composed of a doubly linked list where each node is doiblt linked to the header of a doubly linked list.'''
    def __init__(Net,top=None,icount=0):#,CLL=CLL()):#last and Next are used interchangeably here
        top = Node()
        Net.top          = top;
        Net.bottom       = top # to start header and bottom are in the same place

        Net.top.up       = None
        Net.bottom.down  = None;

        Net.icount = icount;
        Net.top.index = 0;
        # to access the CLL connected to the category
        # if you prefer and/or are actually dealing with netflix type 
        # structure, you cauld call it movies instead
        Net.top.CLL = CLL(header=Node())
        # making sure to account for the index
        Net.top.CLL.icount+=1
        Net.bottom.CLL = Net.top.CLL;
        Net.top.right = Net.top.CLL.header;
        Net.bottom.right = Net.bottom.CLL.header;
    def __add__(Net,category):
      if type(category)==Node:
        category = category.data
      else:
        if Net.top.data == None:# and Net.top.down == None:# do I need the stuff after the and here?
          Net.top.data = category;
        else:
          # make the new category
          Net.bottom.down = Node();
          Net.bottom.down.data = category
          Net.icount+=1;
          Net.bottom.down.index = Net.icount;
          # make and attach a CLL to the category
          Net.bottom.down.CLL = CLL(header=Node())
          Net.bottom.down.CLL.icount+=1
          # make the CLL attribute the same as the right attribute
          Net.bottom.down.right = Net.bottom.down.CLL.header;
          Net.bottom.down.right.left = Net.bottom.down;
          # make the up a down connections between the previous bottom and the new bottom
          Net.bottom.down.up = Net.bottom;
          # move the Net.bottom label to the new bottom
          Net.bottom = Net.bottom.down;
          #Net.bottom.down = None;
      return Net;
    def __str__(Net):
      t = Netflix.top
      k=''
      # print('Categories:')
      k +='Categories:\n'
      while t!=None:
        #print(str(t.data)+':')
        k+=str(t.data)+':\n'
        #print(t,':')# if you want t as the whole node
        #print(t.CLL)
        k+=str(t.CLL)+"\n"
        #print("")
        t=t.down
      return k;
    def at(Net,i):
      temp = Net.top;
      if type(i)==int:
        while temp!=None:
          if (temp.index!=i):
            temp = temp.down;
          else:
            break;
        if temp!=None:
          return temp;
        else:
          return "There is no Node at index "+str(i)+" in this Net.";
      elif type(i)==str:
        while temp!=None:
          if (temp.data!=i or temp.data!=str(i)):
            temp = temp.down;
          else:
            break;
        if temp!=None:
          return temp;
        else:
          return "There is no Node with data "+str(i)+" in this Net.";        
# for Netflix there are categories (DLL) that individually connect to circular linked lists full of movies. There could also be another whole layer
# when comes to movies vs shows vs documentaries vs...? what else is there in Netflix
````
